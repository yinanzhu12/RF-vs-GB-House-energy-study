```python
import pandas as pd
import sklearn.ensemble
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("energydata_complete.csv")
```

```python
data['timeofday']=data['date']
for index in range(0,19735):
    data.loc[index,'timeofday']=int(data.loc[index,'timeofday'][11:13])*60+int(data.loc[index,'timeofday'][14:16])
data
```

```python
data.describe()
```

```python
data_shuffle=data.sample(frac=1).reset_index(drop=True)
```

```python
feature=list(data.columns)
feature.remove('rv1')
feature.remove('rv2')
feature.remove('date')
feature.remove('Appliances')
feature.remove('lights')
feature
```

```python
xtrain=data_shuffle.loc[:13800,feature]
ytrain=data_shuffle.loc[:13800,'Appliances']
xtest=data_shuffle.loc[13801:,feature]
ytest=data_shuffle.loc[13801:,'Appliances']
```

```python
def plotscore(samplescore,testscore,ran):
    plt.figure()
    ss,=plt.plot(ran,samplescore)
    ts,=plt.plot(ran,testscore)
    plt.legend([ss,ts],["sample","test"])
    plt.show()
```

```python
samplescore=[]
testscore=[]
ran=range(2,200,10)
for N in ran:
    rf=sklearn.ensemble.RandomForestRegressor(n_estimators=100,max_features=5,max_leaf_nodes=N)
    rf.fit(xtrain,ytrain)
    samplescore.append(rf.score(xtrain,ytrain))
    testscore.append(rf.score(xtest,ytest))
```

```python
plotscore(samplescore,testscore,ran)
```

```python
rf.feature_importances_
```

```python
plt.figure()
plt.bar(np.arange(len(feature)),rf.feature_importances_)
plt.show()
```

```python
samplescore=[]
testscore=[]
ran=range(2,200,10)
for N in ran:
    rf=sklearn.ensemble.GradientBoostingRegressor(n_estimators=100,max_features=5,max_leaf_nodes=N)
    rf.fit(xtrain,ytrain)
    samplescore.append(rf.score(xtrain,ytrain))
    testscore.append(rf.score(xtest,ytest))
```

```python
plotscore(samplescore,testscore,ran)
```
